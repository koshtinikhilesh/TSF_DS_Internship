###Question : 
How to select optimum number of clusters for the datasets. We will calculate the cluster naumber using hte famous elbow method.

We will calculate inertia.

Inertia: Sum of the squares of the distance of the sample to the nearest clusters. If the number tends to constants, then all the sample relate to their clusters and we will conclude the cluster number.

```
# list down the inertial values with the index number as cluster number
k_val = []

# iterate through 1 to 11  and calculate inertia
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    k_val.append(kmeans.inertia_)



plt.plot(range(1,11), k_val)
plt.xlabel("Number of clusters", fontsize=18)
plt.title("K Value verification", fontsize=18)
plt.ylabel("Sum of Squared Error", fontsize=18)
```

Observations:-  Line represents the sum of squared error, smaller the value of error , more dense and accurate the predictions from the actual values.



<br>
So, In the figure, we can see that value of inertia(sum of squared  error) decreases and becomes steadily after n=3. So, we will take number of clusters as 3.
